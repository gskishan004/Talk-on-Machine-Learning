```python
from sklearn import datasets
# Load the diabetes dataset
diabetes = datasets.load_diabetes()
```

```python
%matplotlib inline
```

```python
# more about the dataset : http://www4.stat.ncsu.edu/~boos/var.select/diabetes.html
# 442 diabetes patients were measured on 10 baseline variables.
# response variable, a measure of disease progression one year after baseline.
# Normalized data
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread("diabetes.PNG")
imgplot = plt.imshow(img)
plt.show()
```

```python
#set of features and desired output
X = diabetes.data
y = diabetes.target
print X.shape
print y.shape
```

```python
for a in range(10):
    print X[0][a]
```

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=1)
print X.shape
print y.shape
```

```python
# Use only one column from data i.e, the second one
print(X.shape)
X = X[:, 2:3]
print(X.shape)
```

```python
# splitting data
train_set_size = 250
X_train = X[:train_set_size]  # selects first 250 rows (examples) for train set
X_test = X[train_set_size:]   # selects from row 250 until the last one for test set
print(X_train.shape)
print(X_test.shape)
```

```python
#splitting targets
y_train = y[:train_set_size]   # selects first 250 rows (targets) for train set
y_test = y[train_set_size:]    # selects from row 250 until the last one for test set
print(y_train.shape)
print(y_test.shape)
```

```python
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
print(y_train.shape)
print(y_test.shape)
```

```python
#visualizing the data
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.xlabel('Data')
plt.ylabel('Target')
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
#fit the model
regr.fit(X_train, y_train);
```

```python
print(regr.coef_)
print(regr.intercept_)
```

```python
# The mean square error
import numpy as np
print("Training error: ", np.mean((regr.predict(X_train) - y_train) ** 2))
print("Test     error: ", np.mean((regr.predict(X_test) - y_test) ** 2))
```

```python
# Visualises dots, where each dot represent a data exaple and corresponding teacher
plt.scatter(X_train, y_train,  color='black')
# Plots the linear model
plt.plot(X_train, regr.predict(X_train), color='blue', linewidth=3);
plt.xlabel('Data')
plt.ylabel('Target')
```
